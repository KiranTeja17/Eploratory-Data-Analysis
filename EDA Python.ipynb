### Feature Engineering

##### Feature 1 - Age

We divide the age into three groups

1. Ages between 0-25 are known as young

2. Ages between 25-50 are known as mid aged

3. Ages above 50 are known as Old

```python
# Dividing the age group into three different groups
def set_age(row):
    if row["Age"] < 25:
        return "Young"
    elif row["Age"] >= 25 and row["Age"] <= 50:
        return "Mid_Aged"
    elif row["Age"] >= 51 and row["Age"] <= 80:
        return "Old"
df = df1.assign(Age_div=df.apply(set_age, axis=1))
df.head()
```

```python
sns.countplot(x = 'Outcome', hue = 'Age_div', data= df)
plt.xlabel('Nature of diabetes')
plt.ylabel('No.of patients')
plt.title('Effect of age on diabetes')
```

Most of the mid aged women are affected with diabetes.

##### Feature 2 : BMI
The BMI is divided into the following categories:


1. Under 18.5 â€“ you are considered underweight and possibly malnourished.
2. 18.5 to 24.9 â€“ you are within a healthy weight range for young and middle-aged adults.
3. 25.0 to 29.9 â€“ you are considered overweight.
4. Over 30 â€“ you are considered obese.

```python
# Defining the BMI and the condtion of the people
def set_bmi(row):
    if row["BMI"] < 18.5:
        return "Under"
    elif row["BMI"] >= 18.5 and row["BMI"] <= 24.9:
        return "Healthy"
    elif row["BMI"] >= 25 and row["BMI"] <= 29.9:
        return "Over"
    elif row["BMI"] >= 30:
        return "Obese"
```

```python
# Adding the new column into the main dataset and assigning the new derived dataset into df
df = df.assign(BMI_derived=df.apply(set_bmi, axis=1))
df.head()
```

```python
# Count plot for BMI_derived
sns.countplot(data=df, x = 'BMI_derived', label='Count')
plt.title('BMI of all women')
plt.xlabel('Body Mass Index')
plt.ylabel('No.of women')
df['BMI_derived'].value_counts()
```

 The following levels of BMI describe the condition of the person.

BMI < 18.5 people are considered underweight and possibly malnourished.

BMI from 18.5 to 24.9 people are within a healthy weight range for young and middle-aged adults.

BMI from 25.0 to 29.9 people are considered overweight.

BMI above 30 people are considered obese.

```python
# Let's see how BMI affect Diabetes 
sns.catplot(x = 'BMI_derived',hue = 'Outcome', data=df, kind="count", height=5,aspect=.8 )
plt.title('Influence of BMI on diabetes')
plt.xlabel('Body Mass Index')
plt.ylabel('No.of women')
```

The most women who are over weight and obese have a very high chance in getting diagnosed with diabetes.

##### Feature 3 - Insulin

If insulin level (2-Hour serum insulin (mu U/ml)) is >= 16 and <= 166, then it is normal range else it is considered as Abnormal

```python
# Defining the range in Insulin and adding that new column to the derived dataset 'df'
def set_insulin(row):
    if row["Insulin"] >= 16 and row["Insulin"] <= 166:
        return "Normal"
    else:
        return "Abnormal"
df = df.assign(INSULIN_der=df.apply(set_insulin, axis=1))
df.head()
```

```python
# Countplot for INSULIN_der
sns.countplot(data=df, x = 'INSULIN_der', label='Count')
plt.title('No.of women with Normal and Abnormal Insulin levels')
plt.xlabel('Insulin Levels')
plt.ylabel('No. of women')
df['INSULIN_der'].value_counts()
```

The no.of women having normal insulin levels are 260 and that of abnormal level of insul are 507

```python
# Let's see how insulin levels affect Diabetes 
sns.catplot(x = 'INSULIN_der',hue = 'Outcome', data=df, kind="count", height=5,aspect=.8 )
plt.title('Influence of Insulin levels on diabetes')
plt.xlabel('Insullin Levels')
plt.ylabel('No.of women')
```

The influence of Abnormal insulin levels is very high on diabetes, women with abnormal insulin levels have around 41% chance of diagnosing with diabetes.

Whereas the women with normal insulin levels have a probability of 0.23 to get diagnosed with diabetes.

```python
# Boxplot of Age and Insulin leveles
sns.boxplot(x = 'Age',y = 'INSULIN_der',hue = 'Outcome', data=df)
plt.title('Influence of Age and Insulin levels on diabetes')
plt.xlabel('Age of the patient')
plt.ylabel('Insulin Levels')
```

From the above plot it is observed that the people within the age 25-45 are more prone to diabetes with normal insulin levels whereas the people with abnormal insulin levels are more prone to diabetes above the age 28.

```python
# Boxplot to find the relation between Age and BMI on diabetes
sns.boxplot(x="Age", y="BMI_derived", hue = "Outcome", data = df)
plt.xlabel('Age of the patient')
plt.ylabel('Body Mass Index')
plt.title('Influence of Age and BMI in diabetes')
```

From the above plot it is observed that patients who are obese as per BMI index are more diabetic in early age of 25 where as patients who are overweight are prone to diabetes in early 30's

There are also some instances where the uneder weight patients are diagnosed with diabetes. This may be because of the lack of nutrients they recieve and also because of weak immune system that is generally observed in the underweight patients.

It is also evident that healthy patients as per BMI index are mostly diabetic above the age of 45 and it can be because of a lot of factors like genetics, unhealthy lifestyle, lack of good diet, no exerrcise, etc,.
